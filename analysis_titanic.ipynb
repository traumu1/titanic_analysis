```python
#############################
### Titanic Data Analysis ###
#############################

# Import libraries
import pandas as pd            # Data handling
import seaborn as sns          # Visualization style
import matplotlib.pyplot as plt # Core plotting

# Upload data (Colab file picker)
from google.colab import files
uploaded = files.upload()

# Load CSV file into a DataFrame
titanic = pd.read_csv("titanic_train.csv")
```

```python
### Basic Data Overview   ###

# Show first rows
print("First 5 rows:")
display(titanic.head())

# Dataset structure and data types
print("\nDataset info:")
titanic.info()

# Missing values per column
print("\nMissing values per column:")
print(titanic.isnull().sum())
```

```python
### Exploratory Analysis   ###

# Overall survival distribution
sns.countplot(x='Survived', data=titanic)
plt.title("Overall Survival Count")
plt.show()

# Survival distribution by gender
sns.countplot(x='Survived', hue='Sex', data=titanic)
plt.title("Survival by Sex")
plt.show()

# Survival distribution by passenger class
sns.countplot(x='Survived', hue='Pclass', data=titanic)
plt.title("Survival by Passenger Class")
plt.show()

# Age distribution split by survival outcome
sns.boxplot(x='Survived', y='Age', data=titanic)
plt.title("Age vs. Survival")
plt.show()
```
